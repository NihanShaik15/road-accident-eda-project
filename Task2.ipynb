```python

# Step 3: Load Dataset
# Loaded the cleaned CSV file into a Pandas DataFrame.

df = pd.read_csv("cleaned_road_accidents.csv")

print("Shape:", df.shape)

df.head()
```

```python
# Step 4: Create SQLite Database
# Created a SQLite database and stored the dataset as a table.

conn = sqlite3.connect("road_accidents.db")

df.to_sql(
    "accidents",
    conn,
    if_exists="replace",
    index=False
)

print("Database Created Successfully")
```

```python
# Step 5: View Table Structure
# Verified the table columns and database schema.
query = """
PRAGMA table_info(accidents);
"""

pd.read_sql(query, conn)
```

```python

# Query 1: Total Accidents
# Calculated the total number of accident records.
query = """
SELECT COUNT(*) AS total_accidents
FROM accidents;
"""

pd.read_sql(query, conn)
```

```python
# Query 2: Top 10 Cities
# Identified the cities with the highest number of accidents.
query = """
SELECT city,
       COUNT(*) AS total_accidents
FROM accidents
GROUP BY city
ORDER BY total_accidents DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
# Query 3: State-wise Analysis
# Analyzed accident counts across different states.
query = """
SELECT state,
       COUNT(*) AS total_accidents
FROM accidents
GROUP BY state
ORDER BY total_accidents DESC;
"""

pd.read_sql(query, conn)
```

```python
# Query 4: Weather Analysis
# Examined the impact of weather conditions on accidents.
query = """
SELECT weather,
       COUNT(*) AS total_accidents
FROM accidents
GROUP BY weather
ORDER BY total_accidents DESC;
"""

pd.read_sql(query, conn)
```

```python
# Query 5: Severity Analysis
# Analyzed the distribution of accident severity levels.$0
query = """
SELECT accident_severity,
       COUNT(*) AS total
FROM accidents
GROUP BY accident_severity
ORDER BY total DESC;
"""

pd.read_sql(query, conn)
```

```python

# Query 6: Weekend Analysis
# Compared accidents occurring on weekends and weekdays.$0
query = """
SELECT is_weekend,
       COUNT(*) AS total
FROM accidents
GROUP BY is_weekend;
"""

pd.read_sql(query, conn)
```

```python
# Query 7: Peak Hour Analysis
# Analyzed accident frequency during peak traffic hours.$0
query = """
SELECT is_peak_hour,
       COUNT(*) AS total
FROM accidents
GROUP BY is_peak_hour;
"""

pd.read_sql(query, conn)
```

```python
# Query 8: Traffic Density Analysis
# Studied the relationship between traffic density and accidents.$0
query = """
SELECT traffic_density,
       COUNT(*) AS total_accidents
FROM accidents
GROUP BY traffic_density
ORDER BY total_accidents DESC;
"""

pd.read_sql(query, conn)
```

```python

# Query 9: Average Casualties Analysis
# Calculated average casualties for each severity level.$0
query = """
SELECT accident_severity,
       ROUND(AVG(casualties),2) AS avg_casualties
FROM accidents
GROUP BY accident_severity;
"""

pd.read_sql(query, conn)
```

```python

# Query 10: Risk Score Analysis
# Identified cities with the highest average risk scores.$0
query = """
SELECT city,
       ROUND(AVG(risk_score),2) AS avg_risk_score
FROM accidents
GROUP BY city
ORDER BY avg_risk_score DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
# Query 11: Road Type Analysis
# Examined accident occurrences across different road types.$0
query = """
SELECT road_type,
       COUNT(*) AS total_accidents
FROM accidents
GROUP BY road_type
ORDER BY total_accidents DESC;
"""

pd.read_sql(query, conn)
```

```python
# Query 12: Vehicles Involved Analysis
# Analyzed the number of vehicles involved in accidents.$0
query = """
SELECT vehicles_involved,
       COUNT(*) AS total_accidents
FROM accidents
GROUP BY vehicles_involved
ORDER BY total_accidents DESC;
"""

pd.read_sql(query, conn)
```

```python
# Query 13: Monthly Accident Trend
# Studied accident trends across different months.$0
query = """
SELECT month,
       COUNT(*) AS total_accidents
FROM accidents
GROUP BY month
ORDER BY total_accidents DESC;
"""

pd.read_sql(query, conn)
```

```python
# Query 14: Dangerous Cities Analysis
# Identified cities with the highest casualty counts.$0
query = """
SELECT city,
       SUM(casualties) AS total_casualties
FROM accidents
GROUP BY city
ORDER BY total_casualties DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
#Highest Severity Cities
# Analyzed average accident severity across cities.$0
query = """
SELECT city,
       ROUND(AVG(accident_severity),2) AS avg_severity
FROM accidents
GROUP BY city
ORDER BY avg_severity DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```
