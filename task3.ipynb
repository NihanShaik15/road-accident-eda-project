```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_theme()
```

```python
df = pd.read_csv("cleaned_road_accidents.csv")

print(df.head())
print(df.shape)
```

```python

#Top Accident Cities
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    y="city",
    order=df["city"].value_counts().index
)

plt.title("Top Accident Cities")
plt.show()
```

```python
#State-wise Accidents
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    y="state",
    order=df["state"].value_counts().index
)

plt.title("State-wise Accident Analysis")
plt.show()
```

```python
#Weather Analysis
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="weather"
)

plt.title("Weather Conditions Analysis")
plt.xticks(rotation=45)
plt.show()
```

```python
#Accident Severity
plt.figure(figsize=(6,6))

df["accident_severity"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Accident Severity Distribution")
plt.ylabel("")
plt.show()
```

```python
#Traffic Density
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="traffic_density"
)

plt.title("Traffic Density Analysis")
plt.show()
```

```python
#Road Type
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="road_type"
)

plt.xticks(rotation=45)
plt.title("Road Type Analysis")
plt.show()
```

```python
#Peak Hour
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="is_peak_hour"
)

plt.title("Peak Hour Analysis")
plt.show()
```

```python
#Weekend Analysis
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="is_weekend"
)

plt.title("Weekend vs Weekday")
plt.show()
```

```python
#Vehicles Involved
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="vehicles_involved"
)

plt.title("Vehicles Involved")
plt.show()
```

```python
#Casualties Distribution
plt.figure(figsize=(8,5))

sns.histplot(
    df["casualties"],
    bins=20,
    kde=True
)

plt.title("Casualties Distribution")
plt.show()
```

```python
#Monthly Trend
monthly = df.groupby("month").size()

plt.figure(figsize=(10,5))

monthly.plot(marker="o")

plt.title("Monthly Accident Trend")
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```
